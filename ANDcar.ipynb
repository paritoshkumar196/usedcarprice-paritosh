```python
import pandas as pd

# Assuming df3 is your DataFrame
def convert_to_float(x):
    x = x.replace('₹', '').replace('Lakh', '').replace(',', '').strip()
    
    if 'Crore' in x:
        # Convert Crore to Lakh (1 Crore = 100 Lakh)
        return float(x.replace('Crore', '').strip()) * 100
    else:
        return float(x)

df3['lakh'] = df3['car_price_in_rupees'].apply(convert_to_float)

# Print unique values in the 'lakh' column                                                                   
unique_lakh_values = df3['lakh'].unique()
print(unique_lakh_values)
```
